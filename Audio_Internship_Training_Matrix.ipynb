```python
import os
import pandas as pd
import soundfile as sf
import librosa

# -----------------------------
# Paths
# -----------------------------
AUDIO_FILE = "khorta.wav"
CSV_FILE = "khorta_Arvind.csv"
OUTPUT_DIR = "khorta_segments"
# Create folder if it does not exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Folder '{OUTPUT_DIR}' is ready.")
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv(CSV_FILE)

# Validate required columns
required_cols = {"start_time", "end_time"}
if not required_cols.issubset(df.columns):
    raise ValueError(f"CSV must contain columns: {required_cols}")

# -----------------------------
# Load audio
# -----------------------------
audio, sr = librosa.load(AUDIO_FILE, sr=None)  # preserve original SR

print(f"Loaded audio: {AUDIO_FILE}")
print(f"Sample rate: {sr} Hz")
print(f"Total duration: {len(audio)/sr:.2f} sec")

# -----------------------------
# Segment audio
# -----------------------------
for idx, row in df.iterrows():
    start_time = float(row["start_time"])
    end_time = float(row["end_time"])

    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)

    segment_audio = audio[start_sample:end_sample]

    # Optional label
    label = row["label"] if "label" in df.columns else "khortha"

    out_filename = f"{label}_{idx:03d}.wav"
    out_path = os.path.join(OUTPUT_DIR, out_filename)

    sf.write(out_path, segment_audio, sr)

print(f"Saved {len(df)} segments to '{OUTPUT_DIR}/'")
```
